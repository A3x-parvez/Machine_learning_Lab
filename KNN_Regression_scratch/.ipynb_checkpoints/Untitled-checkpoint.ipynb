```python
import numpy as np

# Example data
X = np.array([[1], [2], [3], [4], [5]])  # Training data
x = np.array([2.5])  # Test point

# Calculate Euclidean distances
distances = np.sqrt(np.sum((X - x)**2, axis=1))

# Get the indices of the 3 smallest distances
k = 3
indices_of_min_distances = np.argsort(distances)[:k]

# Output the indices
print("Distances:", distances)
print("Indices of 3 nearest neighbors:", indices_of_min_distances)
```
